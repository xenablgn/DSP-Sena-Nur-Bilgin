```python
import numpy as np
import matplotlib.pyplot as plt

# Create a numpy array with shape (10, 3) filled with random integers
random_array = np.random.randint(0, 100, size=(10, 3))
print("My random array:")
print(random_array)

# Visualize the array
plt.figure(figsize=(8, 6))
plt.imshow(random_array, cmap='viridis', interpolation='nearest')
plt.colorbar(label='Random Values')
plt.title('Visualization of Random Arrays')
plt.xlabel('Column')
plt.ylabel('Row')
plt.grid(False)
plt.show()
```
